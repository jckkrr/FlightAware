```python
import re
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

import basicScraper
```

```python
df = pd.DataFrame()

soup = basicScraper.basicScraper('http://www.fallingrain.com/world/AS/airports.html')

main_table = soup.find('table')

trs = main_table.find_all('tr')

# <tr><th>Kind</th> <th>ICAO</th> <th>FAA</th> <th>IATA</th> <th>City</th> <th>Name</th> <th>Latitude</th> <th>Longitude</th> <th>Max Runway</th> <th>Airlines</th></tr>

for tr in trs:
    
    if 'ICAO' not in str(tr):
        
        #print()
        #print(tr)
    
        tds = tr.find_all('td')
        
        '''for i in range(0,2):
            print(tds[i])'''
        
        next_row=df.shape[0]
        
        df.loc[next_row, 'kind'] = tds[0].text
        df.loc[next_row, 'icao'] = tds[1].text
        df.loc[next_row, 'faa'] = tds[2].text
        df.loc[next_row, 'iata'] = tds[3].text
        df.loc[next_row, 'city'] = tds[4].text
        df.loc[next_row, 'name'] = tds[5].text
        df.loc[next_row, 'latitude'] = tds[6].text
        df.loc[next_row, 'longitude'] = tds[7].text
        df.loc[next_row, 'max_runway'] = tds[8].text
        df.loc[next_row, 'airlines'] = tds[9].text
        
df
```

```python
df.to_csv('airport_codes_coords.csv', index=False)
```
