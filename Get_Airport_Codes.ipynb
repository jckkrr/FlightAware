```python
import re
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

import basicScraper
```

```python
soup = basicScraper.basicScraper('https://airportcodes.io/en/country/australia/')

country_section_content = soup.find_all('div', {'class': 'country__section_content rich'})[2]
```

```python
df = pd.DataFrame()

trs = country_section_content.find_all('tr')

for tr in trs:
    if 'ICAO' not in str(tr):
        
        tds = tr.find_all('td')
            
        next_row = df.shape[0]
        
        df.loc[next_row, 'ICAO'] = tds[0].text
        df.loc[next_row, 'IATA'] = tds[1].text
        df.loc[next_row, 'Name'] = tds[2].text
        df.loc[next_row, 'Type'] = tds[3].text
        df.loc[next_row, 'Municipality'] = tds[4].text
        
df
```

```python
df.to_csv('Aus_airport_codes.csv', index=False)
```
