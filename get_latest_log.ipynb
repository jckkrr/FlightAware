```python
import re
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
from datetime import datetime

import latestAirportLog
```

```python
# not an exact week but a good enough approximation

def getWeekOfLogs(icao):
    
    n = 0
    df = latestAirportLog.latestAirportLog(f'https://flightaware.com/live/airport/{icao}/arrivals?;offset={n}')

    if df.shape[0] > 0:    

        while len(set(df['Arrival day'].dropna())) < 7:
                n += 20
                dfx = latestAirportLog.latestAirportLog(f'https://flightaware.com/live/airport/{icao}/arrivals?;offset={n}')
                                
                df = pd.concat([df,dfx], ignore_index=True)
                
                print('.', end='')
                
                if df.shape[0] < 20 or dfx.shape[0] < 10 or n >= 1200:
                    break
                    
                
        now = datetime.now().strftime("%Y%m%d%H%M%S")
        df.to_csv(f'CSVs/{icao}_{now}.csv', index=False)
        print(now)
```

```python
aus_airports = pd.read_csv('Aus_airport_codes.csv')
aus_airports = list(aus_airports['ICAO'])

#aus_airports = ['YBAS','YARA', 'YMEN']

for icao in aus_airports:
    print(icao, end='\t')
    try:
        getWeekOfLogs(icao)
    except:
        print(f'::: fail ::: {icao}')
```
